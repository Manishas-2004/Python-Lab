```python
# 4.Create a histogram for ages of 100 people.

import numpy as np
import matplotlib.pyplot as plt
ages = np.random.randint(1, 100, 100)  # Random ages between 1 and 100
plt.hist(ages, bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Ages')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.show()


```
