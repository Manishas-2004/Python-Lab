```python
# 5.Make a pie chart of daily activities (sleep, study, entertainment, etc.).

import matplotlib.pyplot as plt
activities = ['Sleep', 'Study', 'Entertainment','Social Media', 'Exercise', 'Others']
hours = [8, 6, 4, 2, 4,5]
plt.pie(hours, labels=activities, autopct='%1.1f%%', startangle=140)
plt.title('Daily Activities')
plt.show()
```
