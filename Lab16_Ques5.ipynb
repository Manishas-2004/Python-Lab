```python
# 5.sales comparison of 2 products over a 

import matplotlib.pyplot as plt
import numpy as np
months = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
product1_sales = [150, 200, 180, 220]
product2_sales = [130, 170, 160, 210]
x = np.arange(len(months))
width = 0.35
plt.figure()
plt.bar(x - width/2, product1_sales, width, label='Product 1')
plt.bar(x + width/2, product2_sales, width, label='Product 2')
plt.xlabel("Weeks")
plt.ylabel("Sales")
plt.title("Sales Comparison of 2 Products Over a Month")
plt.xticks(x, months)
plt.legend()
plt.show()
```
