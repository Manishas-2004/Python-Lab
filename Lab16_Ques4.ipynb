```python
# 4.make relation between sales and profit using scatter plot(use grid)

import matplotlib.pyplot as plt
sales = [100, 200, 300, 400, 500, 600]
profit = [20, 40, 60, 80, 100, 120]
plt.figure()
plt.scatter(sales, profit, color='green')
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()
```
