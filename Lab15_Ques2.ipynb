```python
#2.Create a bar chart showing marks in 5 subjects.

import matplotlib.pyplot as plt
subjects = ['Math', 'Science', 'History', 'English', 'Social']
marks = [85, 90, 78, 88, 76]
plt.bar(subjects, marks, color='green')
plt.title('Marks in 5 Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()

```
